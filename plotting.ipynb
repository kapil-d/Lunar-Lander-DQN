```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def moving_average(data, window_size):
    return np.convolve(data, np.ones(window_size) / window_size, mode='valid')

def plot_experiment(filename, algorithm_name, wind_strength, window_size=100, color=None):
    rewards = np.loadtxt(filename)
    
    rewards_moving_avg = moving_average(rewards, window_size)
    episodes = np.arange(len(rewards_moving_avg)) + window_size - 1

    # Plot the moving average
    #plt.figure()
    plt.plot(episodes, rewards_moving_avg, label=f'{algorithm_name}, Wind Speed = {wind_strength}', color=color)
    plt.xlabel('Episode')
    plt.ylabel('Moving Average Reward')

    #plt.show()

    
```

```python
plot_experiment('experiments/env_wind_0_DQN_rewards.txt', 'DQN', 0)
plot_experiment('experiments/env_wind_1_DQN_rewards.txt', 'DQN', 5)
#plot_experiment('experiments/env_wind_2_DQN_rewards.txt', 'DQN', 10)
plot_experiment('experiments/env_wind_3_DQN_rewards.txt', 'DQN', 15, color='tab:red')
plot_experiment('experiments/env_wind_4_DQN_rewards.txt', 'DQN', 20, color='tab:purple')

# plot_experiment('experiments/env_wind_0_DDQN_rewards.txt', 'DDQN', 0, )
# plot_experiment('experiments/env_wind_1_DDQN_rewards.txt', 'DDQN', 5)
# plot_experiment('experiments/env_wind_2_DDQN_rewards.txt', 'DDQN', 10)
# plot_experiment('experiments/env_wind_3_DDQN_rewards.txt', 'DDQN', 15)
# plot_experiment('experiments/env_wind_4_DDQN_rewards.txt', 'DDQN', 20)

# plot_experiment('experiments/env_wind_0_DuelingDQN_rewards.txt', 'Dueling DQN', 0, color='tab:blue')
# #plot_experiment('experiments/env_wind_1_DuelingDQN_rewards.txt', 'Dueling DQN', 5)
# plot_experiment('experiments/env_wind_2_DuelingDQN_rewards.txt', 'Dueling DQN', 10, color='tab:green')
# plot_experiment('experiments/env_wind_3_DuelingDQN_rewards.txt', 'Dueling DQN', 15, color='tab:red')
# plot_experiment('experiments/env_wind_4_DuelingDQN_rewards.txt', 'Dueling DQN', 20, color='tab:purple')



plt.grid()
plt.legend()
plt.show()
```
